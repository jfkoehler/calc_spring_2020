```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
import ipywidgets as widgets
from ipywidgets import interact
```

```python
x = np.linspace(-2, 2, 1000)
```

```python
def f(x):
    return x**2
```

```python
plt.plot(x, f(x))
```

```python
plt.plot(x, f(x))
plt.plot(1, f(1), 'o')
```

```python
def tan_line(x, a):
    return derivative(a)*(x - a) + f(a)

def normal_line(x, a):
    return -1/derivative(a)*(x - a) + f(a)
```

```python
def derivative(a):
    h = 0.0000001
    return (f(a + h) - f(a))/h
```

```python
plt.plot(x, f(x))
plt.plot(1, f(1), 'o')
plt.plot(x, tan_line(x, 1), color = 'orange')
```

```python
plt.plot(x, f(x))
plt.plot(1, f(1), 'o')
plt.plot(x, tan_line(x, 1), color = 'orange')
plt.plot(-1, f(-1), 'o', color = 'orange')
plt.plot(x, tan_line(x, -1), color = 'orange')
```

```python
plt.figure(figsize = (12, 4))
plt.plot(x, f(x))
for p in np.linspace(-2, 2, 20):
    plt.plot(p, f(p), 'o', color = 'orange')
    plt.plot(x, tan_line(x, p), '--', color = 'orange')
plt.ylim(-2, 4)
```

```python
def f(x):
    return x**2
plt.figure(figsize = (12, 6))
plt.plot(x, f(x), color = 'black')
for p in np.linspace(-2, 2, 50):
    plt.plot(p, f(p), 'o', color = 'orange')
    plt.plot(x, normal_line(x, p), color = 'orange')
plt.ylim(-1, 2)
```

```python
def normal_plotter(coef):
    def f(x):
        return coef*x**2
    
    def tan_line(x, a):
        return derivative(a)*(x - a) + f(a)

    def normal_line(x, a):
        return -1/derivative(a)*(x - a) + f(a)
    
    def derivative(a):
        h = 0.0000001
        return (f(a + h) - f(a))/h
    plt.figure(figsize = (12, 6))
    plt.plot(x, f(x), color = 'black')
    for p in np.linspace(-2, 2, 50):
        plt.plot(p, f(p), 'o', color = 'orange')
        plt.plot(x, normal_line(x, p), color = 'orange')
    plt.ylim(-3,3)
```

```python
interact(normal_plotter, coef = widgets.FloatSlider(0.1, min = -2, max = 2))
```
