## Visualizing Differential Equations

$$\frac{df}{dx} = x^2$$

```python
x = np.arange(-4,4)
y = np.arange(-4,4)
X, Y = np.meshgrid(x, y)
plt.plot(X, Y, 'o', color = 'black');
```

$$\frac{dg}{dx} = \sin(x)$$

```python
x = np.arange(-3,3, 0.25)
y = np.arange(-2,2, 0.25)

X,Y = np.meshgrid(x, y)
U = 1
V = np.sin(X)


plt.figure(figsize = (10, 7))
plt.title(r'Solutions to $\displaystyle\frac{df}{dx} = \sin(x)$', fontsize=20, loc = 'left');
plt.quiver(X, Y, U, V)
# for c in np.arange(-3,3):
#     plt.plot(x, -np.cos(x) + c, color = 'blue', label = f'c = {c}')
# plt.legend(fontsize = 20, facecolor = 'aqua', framealpha = 0.99)
plt.ylim(-2,2);
```
