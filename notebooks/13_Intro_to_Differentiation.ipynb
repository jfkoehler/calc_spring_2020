##### Secant



```python
def f(x): return x**2
x = np.linspace(0, 3, 100)

plt.figure(figsize = (15, 5))

plt.plot(x, f(x))

plt.plot([1, 3], [f(1), f(3)], '-o', label = '(0, 3)')

plt.plot([1, 2], [f(1), f(2)], '-o', label = '(0, 2)')

plt.plot([1, 1.1], [f(1), f(1.1)], '-o', label = '(0, 1)')

plt.legend()
```

$$f'(x) = \frac{f(x + h) - f(x)}{h}$$
